# Identifying Names Entities in reggaeton lyrics
In this notebook, I find subsequences of words identified as criminal charges and causes of death in reggaeton lyrics.

### Important note: most of the reggaeton lyrics are written in Spanish

1. First, I read the JSON file generated in step 2 in the "reggaeton_corpus" jupyter notebook and produced a list of lists where every record is a reggaeton song, as we did in step 3 in the same notebook.

```python
with open('regaetton_songs.json', 'r') as file:
    corpus = json.load(file)
```

```python
documents = [d['lyrics'] for artist in corpus.values() if artist for d in artist.values()]
```

2. Then, I run the Stanford CoreNLP server (see details in https://stanfordnlp.github.io/CoreNLP/corenlp-server.html) and get the Name Entity Recognizer annotations for all the corpus. The "ners" variable produced is a dictionary where every key is a class of entities, and every value is a list of subsequences

```python
ners  = util.ners_in_corpus(documents)
```

```python
#These are the classes found in reggaeton corpus:
[k for k in ners.keys()]
```

We can explore the frequencies of the subsequences by class with the function "ners_frequency":

```python
crime = util.ners_frequency(ners, 'CRIMINAL_CHARGE')
crime[:10]
```

```python
cause_death = util.ners_frequency(ners, 'CAUSE_OF_DEATH')
cause_death[:10]
```

```python
organizations = util.ners_frequency(ners, 'ORGANIZATION')
organizations[:10]
```
