3. Then, I create a list containing only the lyrics of the songs. Remember that the dictionary we created previously includes additional information about the songs. For now, we only need the tokens (words). Additionally, I convert all the tokens to lower. It is not necessary, but in my case, I arbitrarily decided to treat words with uppercase equal to lowercase words.

```python
tokens_list = [util.clean_tokens_doc(v['lyrics']) for artist in songs.values() if artist for v in artist.values()]
```

```python
# This is an example of how the list look like:
tokens_list[:10]
```

```python
len(tokens_list)
```

4. The next step is to train the Word2Vec model. I used the Gensim library for simplicity. In my case, I chose 300 as the dimension of the vectors and five as a minimum number of appearances of words. These are the same parameters selected by Rachel Tatman to train a Word2Vec model based on the Spanish Billion Words Corpus (https://www.kaggle.com/rtatman/pretrained-word-vectors-for-spanish). I used those pre-trained vectors for comparison in the "similaritiesw2v" notebook. 
    
You can check the Gensim documentation here: https://radimrehurek.com/gensim/models/word2vec.html#gensim.models.word2vec.Word2Vec

And if you want to know more about this kind of embeddings you can see this famous paper: https://arxiv.org/abs/1301.3781

```python
model = Word2Vec(tokens_list, size=300, min_count=5)
```

```python
words = list(model.wv.vocab)
len(words)
```

5. Once we finished the training part, I store the word vectors to use them in the future. This model did not take a long time to train because the reggaeton corpus is tiny compared to corpora that are usually used to train Word2Vec models. You can see that the corpus produced a vocabulary of only 13,950 words, while the model used for comparison produced 1,000,653 words — this a fundamental limitation to take into account.

```python
word_vectors = model.wv
```

```python
fname = get_tmpfile(os.getcwd()+"\REGGAETON-vectors-300-min5.kv")
word_vectors.save(fname)
word_vectors = KeyedVectors.load(fname, mmap='r')
```
